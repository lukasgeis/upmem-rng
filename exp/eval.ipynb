```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Change to corresponding values
dpu_depth = 11
task_depth = 4

data_file = "../data/exp_{}_{}.csv".format(dpu_depth, task_depth)
fulldata_file = "../data/exp_{}_{}_full.csv".format(dpu_depth, task_depth)

data = pd.read_csv(data_file)
fulldata = pd.read_csv(fulldata_file)
```

```python
for task_exp in range(task_depth + 1):
    nr_task = 1 << task_exp
    sns.lineplot(data = fulldata[fulldata.TASKS == nr_task], x = "DPUS", y = "TIME", hue = "ALGORITHM")
    plt.xscale("log")
    plt.yscale("log") 

    plt.title("{} Tasks".format(nr_task))

    plt.show()

sns.lineplot(data = fulldata, x = "TASKS", y = "TIME", hue = "ALGORITHM")
plt.xscale("log")
plt.yscale("log")
plt.show()
```
